```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
data=load_iris()
```

```python
print(data.DESCR)
```

```python
data.data
```

```python
data.target
```

**Train test split**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(data.data,data.target,test_size=0.3,random_state=18)
X_train.shape,y_train.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier(random_state=18)
dtc
```

```python
dtc.fit(X_train,y_train)
```

```python
y_pred=dtc.predict(X_test)
y_pred
```

```python
from sklearn import tree
tree.plot_tree(dtc,filled=True)
plt.show()
```

```python
dtc=DecisionTreeClassifier(random_state=18,max_depth=2)
dtc.fit(X_train,y_train)
from sklearn import tree
tree.plot_tree(dtc,filled=True)
plt.show()
```

 **Hyper parametertuning**

(*, criterion: str = "gini", splitter: str = "best", max_depth: Any | None = None, min_samples_split: int = 2, min_samples_leaf: int = 1, min_weight_fraction_leaf: float = 0, max_features: Any | None = None, random_state: Any | None = None, max_leaf_nodes: Any | None = None, min_impurity_decrease: float = 0, class_weight: Any | None = None, ccp_alpha: float = 0) -> None

```python
params={'criterion': ["gini", "entropy", "log_loss"],
        'splitter':["best", "random"],
        'max_depth':[2,3,4],
        'min_samples_split': [2,3,4],
        'min_samples_leaf': [1,2,3],
        'min_weight_fraction_leaf': [0,1,2],
        'max_features': [2,3,4],
        'max_leaf_nodes':[2,3,4]

}
params
```

```python
dtc=DecisionTreeClassifier()
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
grid=GridSearchCV(estimator=dtc,param_grid=params,scoring='accuracy')
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred=grid.predict(X_test)
y_pred
```

```python
best=grid.best_estimator_
best
```

```python
tree.plot_tree(best)
plt.show()
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
rand=RandomizedSearchCV(estimator=dtc,param_distributions=params,scoring='accuracy')
rand.fit(X_train,y_train)
y_pred=rand.predict(X_test)
best=rand.best_estimator_
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
best=rand.best_estimator_
tree.plot_tree(best,filled=True)
plt.show()
```
