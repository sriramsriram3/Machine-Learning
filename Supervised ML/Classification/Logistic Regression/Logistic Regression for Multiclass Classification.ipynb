```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
print(iris.DESCR)
```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
iris.data
```

```python
iris.target
```

```python
df=pd.DataFrame(iris.data)
df
```

```python
df['Flower']=pd.DataFrame(iris.target)
df.head()
```

**Train and test split**

```python
X=df.iloc[:,:-1]
X.shape
```

```python
y=df.iloc[:,-1]
y
```

```python
X.shape,y.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=45)
X_train.shape,X_test.shape
```

**Feature engineering and Cross Validation**

```python
from sklearn.model_selection import StratifiedKFold
cv=StratifiedKFold(shuffle=True,random_state=45)
```

**Model development**

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(multi_class='ovr')
lr
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
y_pred
```

```python
lr.predict_proba(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
cm=confusion_matrix(y_test,y_pred)
print('confusion matrix \n', cm)
ass=accuracy_score(y_test,y_pred)
print('accuracy' , ass)
ps=precision_score(y_test,y_pred,average='macro')
print('precission',ps)
rec=recall_score(y_test,y_pred,average='macro')
print('recall',rec)
f1=f1_score(y_test,y_pred,average='micro')
print('f1 score',f1)
```

# **Grid Search CV**

Hyper parameter tuning

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(multi_class='ovr')
```

```python
Solver=['lbfgs', 'newton-cholesky', 'newton-cg', 'sag', 'saga', 'liblinear']
Penalty=['l1', 'l2','elasticnet']
c=[0.001,0.01,0.005,0.05,1,10,100]
max_iter= [100, 200, 300]
params=dict(solver=Solver,penalty=Penalty,C=c)
params
```

```python
grid=GridSearchCV(estimator=lr,param_grid=params,scoring='accuracy',cv=cv)
grid
```

```python
grid.fit(X_train,y_train)
```

```python
y_pred=grid.predict(X_test)
y_pred
```

```python
grid.best_estimator_
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
cm=confusion_matrix(y_test,y_pred)
print('confusion matrix \n', cm)
ass=accuracy_score(y_test,y_pred)
print('accuracy' , ass)
ps=precision_score(y_test,y_pred,average='macro')
print('precission',ps)
rec=recall_score(y_test,y_pred,average='macro')
print('recall',rec)
f1=f1_score(y_test,y_pred,average='macro')
print('f1 score',f1)
```

# **RandomSearchCV**

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression(multi_class='ovr')
```

```python
Solver=['lbfgs', 'newton-cholesky', 'newton-cg', 'sag', 'saga', 'liblinear']
Penalty=['l1', 'l2','elasticnet']
c=[0.001,0.01,0.005,0.05,1,10,100]
max_iter= [100, 200, 300]
params=dict(solver=Solver,penalty=Penalty,C=c,max_iter=max_iter)
params
```

```python
rand=RandomizedSearchCV(estimator=lr,param_distributions=params,scoring='accuracy',cv=cv,random_state=45,n_jobs=-1)
rand
```

```python
rand.fit(X_train,y_train)
```

```python
y_pred=rand.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
cm=confusion_matrix(y_test,y_pred)
print('confusion matrix \n', cm)
ass=accuracy_score(y_test,y_pred)
print('accuracy' , ass)
ps=precision_score(y_test,y_pred,average='macro')
print('precission',ps)
rec=recall_score(y_test,y_pred,average='macro')
print('recall',rec)
f1=f1_score(y_test,y_pred,average='macro')
print('f1 score',f1)
```
