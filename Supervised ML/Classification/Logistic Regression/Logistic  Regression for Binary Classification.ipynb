```python
import numpy as  np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_classification
X,y=make_classification(n_samples=1000,n_features=10,n_classes=2,random_state=30)
X.shape,y.shape
```

```python
df=pd.DataFrame(X)
df.head()
```

```python
y
```

# Train trst spllitting

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=30,random_state=35)
X_train.shape,y_train.shape
```

# Model training

```python
from sklearn.linear_model import LogisticRegression
logistic=LogisticRegression()
logistic.fit(X_train,y_train)
```

# Prediction on test data

```python
y_pred=logistic.predict(X_test)
y_pred
```

```python
plt.scatter(y_test,y_pred)
plt.show()
```

```python
logistic.predict_proba(X_test)
```

# Performance Metrics

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
cm=confusion_matrix(y_test,y_pred)
print('confusion matrix \n', cm)
ass=accuracy_score(y_test,y_pred)
print('accuracy' , ass)
ps=precision_score(y_test,y_pred)
print('precission',ps)
f1=f1_score(y_test,y_pred)
print('f1 score',f1)
```

#                              **Hyper parameter tuning**

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
from sklearn.linear_model import LogisticRegression
```

```python
penalty=['l1', 'l2']
solver=[ 'liblinear', 'saga']
c=[0.1,0.01,0.001,10,100]
parms=dict(penalty=penalty,solver=solver,C=c)
parms
```

```python
model=LogisticRegression()
cv=StratifiedKFold()
grid=GridSearchCV(estimator=model,param_grid=parms,cv=cv,n_jobs=-1,scoring='accuracy')
```

```python
grid
```

```python
grid.fit(X_train,y_train)
```

```python
y_pred=grid.predict(X_test)
y_pred
```

```python
grid.best_score_
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
cm=confusion_matrix(y_test,y_pred)
print('confusion matrix \n', cm)
ass=accuracy_score(y_test,y_pred)
print('accuracy' , ass)
ps=precision_score(y_test,y_pred)
print('precission',ps)
f1=f1_score(y_test,y_pred)
print('f1 score',f1)
```

# **Randomized Search CV**

```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import StratifiedKFold
from sklearn.linear_model import LogisticRegression
```

```python
penalty=['l1', 'l2','elasticnet']
solver=[ 'lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag']
c=[0.1,0.01,0.001,10,100]
parms=dict(penalty=penalty,solver=solver,C=c)
parms
```

```python
rand=RandomizedSearchCV(estimator=model,param_distributions=parms,scoring='accuracy')
```

```python
rand
```

```python
rand.fit(X_train,y_train)
```

```python
y_pred=rand.predict(X_test)
```
