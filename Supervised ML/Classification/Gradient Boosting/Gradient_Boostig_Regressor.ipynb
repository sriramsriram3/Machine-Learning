```python
X=df.drop('Collection',axis=1)
y=df['Collection']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=18,shuffle=True)
X_train.shape,X_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet, LogisticRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor,GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor

# Dictionary of regression models
regression_models = {
    'Linear Regression': LinearRegression(),
    'Ridge Regression': Ridge(),
    'Lasso Regression': Lasso(),
    'ElasticNet Regression': ElasticNet(),
    'Decision Tree Regressor': DecisionTreeRegressor(),
    'Random Forest Regressor': RandomForestRegressor(),
    'Support Vector Regressor': SVR(),
    'K-Nearest Neighbors Regressor': KNeighborsRegressor(),
    'Ada boosting Regressor' : AdaBoostRegressor(),
    'Logistic Regression' : LogisticRegression(),
    'Gradient regression': GradientBoostingRegressor()
}
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score, median_absolute_error

def performance(y_test, y_pred):
    mse = mean_squared_error(y_test, y_pred)
    rmse = mean_squared_error(y_test, y_pred, squared=False)
    mae = mean_absolute_error(y_test, y_pred)
    medae = median_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    print(f"Mean Squared Error (MSE): {mse:.4f}")
    print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
    print(f"Mean Absolute Error (MAE): {mae:.4f}")
    print(f"Median Absolute Error: {medae:.4f}")
    print(f"R-squared (R2): {r2:.4f}")
```

```python
for name,models in regression_models.items():
  print(name)
  model=models
  model.fit(X_train,y_train)
  y_pred=model.predict(X_test)
  performance(y_test,y_pred)
```

```python
from sklearn.model_selection import RandomizedSearchCV
for name,method,hyperparams in regression_models:
  rand=RandomizedSearchCV(estimator=method,scoring='accuracy',param_distributions=hyperparams,n_jobs=-1,random_state=18)
  print(name)
  rand.fit(X_train,y_train)
  y_pred=rand.predict(X_test)
  performance(y_test,y_pred)


```
